```python

 import numpy as np
import cv2
from scipy.spatial import distance
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical

import tensorflow as tf
import matplotlib.pyplot as plt

# Define the function to calculate the Eye Aspect Ratio (EAR)
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])  # Vertical distance
    B = distance.euclidean(eye[2], eye[4])  # Vertical distance
    C = distance.euclidean(eye[0], eye[3])  # Horizontal distance
    ear = (A + B) / (2.0 * C)  # EAR formula
    return ear

# Example of facial landmark points for left and right eyes (replace with actual landmarks)
left_eye_landmarks = np.array([[36, 48], [37, 50], [38, 52], [39, 50], [40, 48], [41, 46]])
right_eye_landmarks = np.array([[42, 48], [43, 50], [44, 52], [45, 50], [46, 48], [47, 46]])

# Calculate EAR for both eyes
left_ear = eye_aspect_ratio(left_eye_landmarks)
right_ear = eye_aspect_ratio(right_eye_landmarks)
average_ear = (left_ear + right_ear) / 2.0

# Print EAR results
print(f"Left EAR: {left_ear:.2f}")
print(f"Right EAR: {right_ear:.2f}")
print(f"Average EAR: {average_ear:.2f}")

# Example threshold and drowsiness detection logic
thresh = 0.25
if average_ear < thresh:
    print("*ALERT!* Drowsiness Detected")
else:
    print("Eyes Open - All Good")

# Define the teacher model
def create_teacher_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 1)),  # Example input shape
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(2, activation='softmax')  # Output layer for binary classification
    ])
    return model

# Define the student model
def create_student_model():
    model = Sequential([
        Conv2D(16, (3, 3), activation='relu', input_shape=(64, 64, 1)),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(2, activation='softmax')  # Output layer for binary classification
    ])
    return model

# Distillation loss function
def distillation_loss(y_true, y_pred, teacher_logits, temperature=3):
    y_true = tf.convert_to_tensor(y_true, dtype=tf.float32)
    teacher_logits = tf.convert_to_tensor(teacher_logits, dtype=tf.float32)
    teacher_probs = tf.nn.softmax(teacher_logits / temperature)
    student_probs = tf.nn.softmax(y_pred / temperature)
    distillation_loss = tf.reduce_mean(tf.keras.losses.kullback_leibler_divergence(teacher_probs, student_probs))
    return distillation_loss

# Create and compile the teacher model
teacher_model = create_teacher_model()
teacher_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Assuming we have some training data (X_train, y_train)
# Here, use random data for demonstration; replace it with your actual data
X_train = np.random.rand(100, 64, 64, 1)  # Example input data
y_train = to_categorical(np.random.randint(2, size=100), num_classes=2)  # Example labels

# Train the teacher model (replace with actual data training)
history_teacher = teacher_model.fit(X_train, y_train, epochs=10, batch_size=10)

# Get the teacher model's predictions on the training data
teacher_logits = teacher_model.predict(X_train)

# Create and compile the student model
student_model = create_student_model()
student_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# To store distillation loss
distillation_losses = []

# Train the student model with knowledge distillation
for epoch in range(10):  # Number of epochs for the student
    print(f"Epoch {epoch + 1}/10")
    
    # Train the student model
    history_student = student_model.fit(X_train, y_train, epochs=1, batch_size=10, verbose=0)

    # Calculate distillation loss
    preds = student_model.predict(X_train)
    loss = distillation_loss(y_train, preds, teacher_logits).numpy()
    distillation_losses.append(loss)
    print(f"Distillation Loss: {loss:.4f}")

# Evaluate the student model
test_loss, test_acc = student_model.evaluate(X_train, y_train)
print(f"Test Accuracy of Student Model: {test_acc * 100:.2f}%")

# Save the models for later use
teacher_model.save('drowsiness_detection_teacher_model.h5')
student_model.save('drowsiness_detection_student_model.h5')

# Plotting performance graphs
# 1. Training Loss for Teacher Model
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_teacher.history['loss'], label='Teacher Loss', color='blue')
plt.title('Training Loss vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# 2. Training Accuracy for Teacher Model
plt.subplot(1, 2, 2)
plt.plot(history_teacher.history['accuracy'], label='Teacher Accuracy', color='blue')
plt.title('Training Accuracy vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Plotting student model performance
plt.figure(figsize=(12, 5))
# 3. Distillation Loss for Student Model
plt.plot(distillation_losses, label='Distillation Loss', color='red')
plt.title('Distillation Loss vs. Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

```
